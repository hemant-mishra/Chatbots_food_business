```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report
from sklearn.linear_model import LogisticRegression
```

```python
df=pd.read_csv(r"/content/Bitext_Sample_Customer_Service_Training_Dataset.csv")
```

```python
df["category"].unique()
```

```python
df["intent"].nunique()
```

```python
df1=df.groupby(["intent"]).count()
```

```python
df1
```

```python
import nltk
try:
  nltk.data.find("tokenizer/punkt")
except:
  nltk.download("punkt")
```

```python
label_intent=preprocessing.LabelEncoder()
df["intent_num"]=label_intent.fit_transform(df["intent"])
df.head()
```

```python
x_train,x_test,y_train,y_test=train_test_split(df["utterance"],df["intent_num"],test_size=0.2,random_state=2022,stratify=df["intent_num"])
```

```python
label_intent.__dict__
```

```python
print(x_train.shape)
print(y_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
clf_tfidf=Pipeline([("Vectorizer_tfid",TfidfVectorizer()),("log_classifier",LogisticRegression(C=1.0,penalty="l2",max_iter=5))])
```

```python
clf_tfidf.fit(x_train,y_train)
```

```python
y_pred=clf_tfidf.predict(x_test)
```

```python
print(classification_report(y_test,y_pred))
```

```python
from sklearn.feature_extraction.text import CountVectorizer
clf_bow = Pipeline([("Vectorizer_tfid",CountVectorizer()),("log_classifier",LogisticRegression(C=1.0,penalty="l2",max_iter=5))])
```

```python
clf_bow.fit(x_train,y_train)
```

```python
y_pred1=clf_tfidf.predict(x_test)
```

```python
print(classification_report(y_test,y_pred1))
```
